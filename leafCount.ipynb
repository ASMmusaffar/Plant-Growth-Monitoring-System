# Leaf Count For No-Controls and Water-Control

This script helps in tracking and visualizing plant growth by analyzing images taken over a period of time. By detecting the green areas in the images, it provides a quantitative measure of growth, which is then visualized through a line plot

## No-Controls

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

# Set the folder path containing images
folder_path = "PlantSelected/NoControls"  # Change this to your folder path

# List all image files in the folder
image_files = [f for f in os.listdir(folder_path) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Store results
leaf_counts = {}
min_contour_area = 800  # Adjust based on leaf size

# Loop through all images
for image_file in image_files:
    image_path = os.path.join(folder_path, image_file)
    image = cv2.imread(image_path)

    if image is None:
        print(f"Failed to load image: {image_file}")
        continue

    # Resize for consistency
    scale_percent = 50
    width = int(image.shape[1] * scale_percent / 100)
    height = int(image.shape[0] * scale_percent / 100)
    image = cv2.resize(image, (width, height), interpolation=cv2.INTER_AREA)

    # Convert to HSV and apply threshold
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    lower_green = np.array([30, 40, 40])
    upper_green = np.array([90, 255, 255])
    mask = cv2.inRange(hsv, lower_green, upper_green)

    # Morphological operations
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
    mask_cleaned = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    mask_cleaned = cv2.morphologyEx(mask_cleaned, cv2.MORPH_OPEN, kernel)

    # Distance transform and watershed
    dist_transform = cv2.distanceTransform(mask_cleaned, cv2.DIST_L2, 5)
    _, sure_foreground = cv2.threshold(dist_transform, 0.5 * dist_transform.max(), 255, 0)
    sure_foreground = np.uint8(sure_foreground)

    _, markers = cv2.connectedComponents(sure_foreground)
    markers = markers + 1  # Background labeled as 1

    # Apply watershed
    cv2.watershed(cv2.cvtColor(mask_cleaned, cv2.COLOR_GRAY2BGR), markers)
    image[markers == -1] = [0, 0, 255]  # Red boundary

    # Find contours and count leaves
    contours, _ = cv2.findContours(sure_foreground, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    leaf_count = 0

    for contour in contours:
        area = cv2.contourArea(contour)
        if area > min_contour_area:
            hull = cv2.convexHull(contour)
            hull_area = cv2.contourArea(hull)
            solidity = float(area) / hull_area if hull_area > 0 else 1

            if solidity > 0.7:
                leaf_count += 1

    # Store the count for graphing
    leaf_counts[image_file] = leaf_count
    print(f"{image_file}: {leaf_count} leaves detected")

# Plot the results
plt.figure(figsize=(10, 5))
plt.bar(leaf_counts.keys(), leaf_counts.values(), color='green')
plt.xlabel("Image File")
plt.ylabel("Leaf Count")
plt.title("Leaf Count Per Image For No-Control")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
